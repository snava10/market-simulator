```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import plot

def plot_freq(pdf):    
    freq = np.fft.fft(pdf['price'].values)
    fig = plot(np.abs(freq))
    plt.show(fig)

df = pd.read_csv("../ml-finance/data/btc.csv", names=['date', 'price'], parse_dates=True, index_col='date')
df_norm = (df - df.mean())/df.max()
count = df.shape[0]
buckets = []
for i in range(0, count, 200):
    buckets.append(df_norm.iloc[i:min(i+200,count)])

plot(buckets[0])
plt.show()
plot_freq(buckets[0])

#fig2 = plot(df_norm)
#plt.show(fig2)

prices = df_norm['price'].values
freqs = np.fft.fft(prices)

fig3 = plot(np.abs(freqs))
plt.show(fig3)

```
